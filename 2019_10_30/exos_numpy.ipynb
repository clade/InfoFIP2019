```python
# Diagramme de bode
import matplotlib.pyplot as plt
import numpy as np

def H(omega, omega_0=2*np.pi*1000, zeta=0.1):
    return (omega_0**2/(omega_0**2 - omega**2 + 2j * zeta * omega * omega_0))

pulsations = np.logspace(2, 4, 101) * 2 * np.pi
Hcalc = H(pulsations)

plt.subplot(2, 1, 1)
plt.loglog(pulsations / (2*np.pi), np.abs(Hcalc), '-.k', label='Amplitude')
plt.legend()
plt.title('Oscillateur amorti ($\omega_0/2\pi = 1000$ Hz et $\zeta=0.1$)')
plt.grid(which = 'both')
plt.ylabel('Amplitude')

plt.subplot(2, 1, 2)
plt.semilogx(pulsations / (2*np.pi), 180 / np.pi * np.angle(Hcalc), color='k',linewidth=.75)
plt.grid(which = 'both')
plt.xlabel('Fréquence [Hz]')
plt.ylabel('Phase [°]')

#plt.savefig('Graphiques.svg') 
```

```python
# Integrale de Simpson

def Simpson(f, a, b, N):
    dx = (b-a)/N
    x = a + np.arange(N)*dx
    return np.sum(f(x) + 4*f(x+dx/2) + f(x+dx))*dx/6

print(Simpson(lambda x:1/(1+x**2), 0, 1, 10) - np.pi/4)
print(Simpson(lambda x:1/(1+x**2), 0, 1, 100) - np.pi/4)
```
